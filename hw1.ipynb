Data Cleaning

```python
import pandas as pd
```

Loading the CSV in pandas

```python
data=pd.read_csv('Uncleaned_DS_jobs.csv')
```

```python
data.head()
```

Removing index column

```python
data.drop(columns='index', inplace=True,axis=1)
```

```python
data.head()
```

Checking Errors in the Salary Estimate Column

```python
data['Salary Estimate']
```

```python
data['Salary Estimate'].isna()
```

```python
data['Salary Estimate'].isna().sum()
```

```python
data['Salary Estimate'].isnull().sum()
```

```python
data['Job Description']=data['Job Description'].astype(str).str.strip("\n")
```

Removal of \n from the data in Job Description column

```python
data['Job Description'] = data['Job Description'].astype(str).str.replace("\n", " ")
```

Dropping the rating columns

```python
data.drop(columns='Rating',inplace=True,axis=1)
```

```python
data.head()
```

Removing the \nX.x from the Company name columns

```python
data['Company Name']=data['Company Name'].str.split("\n", n=1).str[0]
```

```python
data.head()
```

Removing the all the demoninations("K","$","Glassdppr est.") from the Salary Estimate column

```python
data['Salary Estimate'] = data['Salary Estimate'].replace({'\$': '', 'K': '', ' \(Glassdoor est.\)': '','(Employer est.)':''}, regex=True)
```

```python
data.head()
```

Checking for Errors in Location Column

```python
data['Location'].isnull().sum()
```

Checking for Errors in Headqaurters Column

```python
data['Headquarters'].isnull().sum()
```

Replacing Invalid values by NA in Size Column

```python
data['Size'].unique()
```

```python
data['Size']=data['Size'].replace(['-1','Unknown'],pd.NA)
```

Replacing Invalid values by NA in Founded Column

```python
data['Founded'].unique()
```

```python
data['Founded']=data['Founded'].replace([-1],pd.NA)
```

```python
data['Founded']
```

Replacing Invalid values by NA in Type of Ownership Column

```python
data['Type of ownership'].unique()
```

```python
data['Type of ownership']=data['Type of ownership'].replace(['-1','Unknown'],pd.NA)
```

Replacing Invalid values by NA in Type of Industry Column

```python
data['Industry'].unique()
```

```python
data['Industry']=data['Industry'].replace(['-1'],pd.NA)
```

```python
data['Industry'].unique()
```

Replacing Invalid values by NA in Sector Column

```python
data['Sector'].unique()
```

```python
data['Sector']=data['Sector'].replace(['-1'],pd.NA)
```

Replacing Invalid values by NA in Revenue Column

```python
data['Revenue'].unique()
```

```python
data['Revenue']=data['Revenue'].replace(['-1'],pd.NA)
```

```python
data['Revenue']
```

```python
data['Competitors'].unique()
```

```python
data['Competitors']=data['Competitors'].replace(['-1'],pd.NA)
```

```python
data.head()
```

Extraction of Salary Estimate data in Mininum, Maximum and Average Salaries

```python
data['Salary Estimate']=data['Salary Estimate'].str.extract('(\d+-\d+)')
```

```python
data[['Salary Min', 'Salary Max']] = data['Salary Estimate'].str.split('-', expand=True).apply(pd.to_numeric, errors='coerce')
```

```python
data['Salary Avg']=(data['Salary Min']+data['Salary Max'])/2
```

```python
data.head()
```

Remaning of Salary Columns with proper Denomination

```python
data.rename(columns={'Salary Estimate':'Salary Range in $ per K','Salary Min':'Salary Minimum in $ per K','Salary Max':'Salary Maximum in $ per K','Salary Avg':'Average Salary in $ per K'},inplace=True)
```
